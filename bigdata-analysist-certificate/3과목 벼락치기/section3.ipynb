### 3-1-b

t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)

```python
means = df['height'].mean()
```

```python
import scipy.stats
```

```python
from scipy.stats import t
```

```python
std = df['height'].std()
```

```python
n = len(df.height)
confidence_level = 0.95
ddof = n - 1
```

```python
t_value = round(t.ppf((1 + confidence_level) / 2, ddof),4)
print(t_value)
```


### 3-1-c

95% 신뢰구간을 구하여라(print(lower,upper) 방식으로 출력, 각각의 값은 소숫점 이하 3째자리까지)

```python
lower = round(means - t_value * std / np.sqrt(n), 3)
upper = round(means + t_value * std / np.sqrt(n), 3)
print(lower, upper)
```

### 3-1-c

카이제곱검정 독립성 검정의 pvalue를 소숫점 이하 3자리까지 구하여라. 유의수준 0.05하에서 귀무가설과 대립가설중 유의한 것을 출력하라


어느 학교에서 수학 시험을 본 학생 100명 중 60명이 60점 이상을 받았다. 이 학교의 수학 시험의 평균 점수가 50점 이상인지 95%의 신뢰 수준에서 검정하려한다.

```python
from scipy.stats import ttest_1samp
```

```python
import numpy as np
n = 100
p_hat = 0.6
p = 0.5
alpha = 0.05

# 검정 통계량 계산
z = round((p_hat - p) / np.sqrt(p * (1 - p) / n),5)
```

```python
from scipy.stats import norm
p_value = round(1 - norm.cdf(z),3)

print(p_value)
print('대립')
```

투약 후 체중에서 투약 전 체중을 뺏을 때 값은 일반 적으로 세가지 등급으로 나눈다. -3이하 : A등급, -3초과 0이하 : B등급, 0 초과 : C등급. 약 실험에서 A,B,C 그룹간의 인원 수 비율은 2:1:1로 알려져 있다. 위 데이터 표본은 각 범주의 비율에 적합한지 카이제곱 검정하려한다.

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_1.csv')
df.head()
```

```python
df['차이'] = df['투약후'] - df['투약전']
```

```python
def china(x):
  if x <= -3:
    return 'A'
  elif -3 < x <= 0:
    return 'B'
  else:
    return 'C'

df['클래스'] = df['차이'].apply(china)
```

```python
df
```
