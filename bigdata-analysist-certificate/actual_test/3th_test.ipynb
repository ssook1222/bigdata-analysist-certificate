## 작업 2유형

### 패키지 다 가져오기

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

### 데이터 다 가져오기

```python
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_test_.csv')
```

```python
print(train.columns)
```

```python
x_train = train[['Age', 'Employment Type', 'GraduateOrNot', 'AnnualIncome',
       'FamilyMembers', 'ChronicDiseases', 'FrequentFlyer',
       'EverTravelledAbroad']]

x_test = test[['Age', 'Employment Type', 'GraduateOrNot', 'AnnualIncome',
       'FamilyMembers', 'ChronicDiseases', 'FrequentFlyer',
       'EverTravelledAbroad']]
```

```python
y_train = train[['ID', 'TravelInsurance']]
```

### 데이터 전처리하기

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
print(x_train.info())
print(x_test.info())
```

```python
columns = ['Employment Type', 'GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad']
```

```python
le = LabelEncoder()
```

```python
for column in columns:
  x_train[column] = le.fit_transform(x_train[column])
  x_test[column] = le.fit_transform(x_test[column])
```

### 데이터 모델링

```python
rfc = RandomForestClassifier()
model = rfc.fit(x_train, y_train)
predict1 = model.predict_proba(x_test)
```

### 데이터 제출

```python
ans = []
for pre in predict1[1]:
  pre = str(pre)[1:-1]
  print(pre)
  raw = list(pre.split())
  ans.append(raw[0])
```

```python
y_columns = ['ID', 'proba']
datas = {'ID': test['ID'], 'proba': ans}

submit = pd.DataFrame(data = datas, columns = y_columns)
submit.head(5)
```

```python
submit.to_csv('result.csv')
```

## 작업형 3영역

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_1.csv')
df.head()
```

다이어트약의 전후 체중 변화 기록이다.     
투약 후 체중에서 투약 전 체중을 뺏을 때 값은 일반 적으로 세가지 등급으로 나눈다.
-3이하 : A등급, -3초과 0이하 : B등급, 0 초과 : C등급.   
약 실험에서 A,B,C 그룹간의 인원 수 비율은 2:1:1로 알려져 있다. 위 데이터 표본은 각 범주의 비율에 적합한지 카이제곱 검정하려한다.

```python
df['difference'] = df['투약후'] - df['투약전']
```

```python
def classify(x):
  if x <= -3:
    return 'A'

  elif -3 < x <= 0:
    return 'B'

  else:
    return 'C'

df['class'] = df['difference'].apply(classify)
```

```python
print(len(df[(df['class'] == 'A')]))
```

### 3-1-b

카이제곱검정 통계량을 반올림하여 소숫점 이하 3째자리까지 구하여라

다시 한 번 풀어보기!

```python
from scipy.stats import chisquare
```

```python
target = df['class'].value_counts().sort_index().to_frame()
target['expected'] = [target['class'].sum()*0.5,target['class'].sum()*0.25,target['class'].sum()*0.25]
```

```python
s,p = chisquare(target['class'],target['expected'])
```

```python
print(round(s, 3))
```

### 3-1-c

카이제곱 검정 p값을 반올림하여 소숫점 이하 3자리까지 구하고, 유의수준 0.05하에서 귀무가설과 대립가설중 유의한 가설을 하나를 선택하시오(귀무/대립)

```python
round_p = round(p, 3)

print(round_p)
if round_p <= 0.05:
  print('대립가설')

else:
  print('귀무가설')
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_2_.csv')
df.head()
```

A, B 공장에서 생산한 기계들의 rpm 값들을 기록한 데이터이다.   
대응 표본 t 검정을 통해 B공장 제품들이 A 공장 제품들보다 rpm이 높다고 말할 수 있는지 검정하려한다.

### 3-2-a

A,B 공장 각각 정규성을 가지는지 샤피로 검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

```python
from scipy.stats import shapiro
```

```python
a_class = df[(df.group == 'A')].rpm
b_class = df[(df.group == 'B')].rpm
```

```python
a_num = shapiro(a_class).pvalue
b_num = shapiro(b_class).pvalue
```

```python
if a_num >= 0.05:
  print("a_class : %f 으로 유의수준 내에서 귀무가설을 기각하므로 정규성을 만족한다." %a_num)
elif a_num < 0.05:
  print("a_class : %f 으로 유의수준 내에서 귀무가설을 채택하므로 정규성을 만족 못한다." %a_num)

if b_num >= 0.05:
  print("b_class : %f 으로 유의수준 내에서 귀무가설을 기각하므로 정규성을 만족한다." %b_num)
elif b_num < 0.05:
  print("b_class : %f 으로 유의수준 내에서 귀무가설을 채택하므로 정규성을 만족 못한다." %b_num)
```

### 3-2-b

A,B 공장 생산 제품의 rpm은 각각 정규성을 가지는지 levene 검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

```python
from scipy.stats import levene
```

```python
s, p = levene(a_class, b_class)
```

```python
print(p)
```

```python
if p >= 0.05:
  print("%f 으로 유의수준 내에서 귀무가설을 기각하므로 정규성을 만족한다." %p)
elif p < 0.05:
  print("%f 으로 유의수준 내에서 귀무가설을 채택하므로 정규성을 만족 못한다." %p)
```

### 3-2-c

대응 표본 t 검정을 통해 B공장 제품들의 rpm이 A 공장 제품의 rpm보다 크다고 말할 수 있는지 검정하라. pvalue를 소숫점 이하 3자리까지 출력하고 귀무가설, 대립가설 중 하나를 출력하라

! 다시 풀어보기

```python
from scipy.stats import ttest_rel
s,p =ttest_rel(b_class,a_class,alternative='greater')
round_p = round(p,3)

print(round_p)
print('대립')
```
