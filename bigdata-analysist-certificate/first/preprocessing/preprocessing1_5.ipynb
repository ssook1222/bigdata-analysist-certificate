```python
import pandas as pd

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)
```

```python
df_raw = df[['quantity', 'item_price']]
df['new_price'] = df_raw['item_price'].str[1:].astype('float')
```

### 41. df의 데이터 중 item_name 값이 N으로 시작하는 데이터를 모두 추출하라

```python
df[df.item_name.str.startswith('N')]
```

### 42. df의 데이터 중 item_name 값의 단어갯수가 15개 이상인 데이터를 인덱싱하라

```python
df[df.item_name.str.len() >=15]
```

조건을 맞춘 행을 찾을 때에는 `df['열 이름']`보다는 `df.열 이름.`..     
이걸로 맞춰서 진행하자!

### 43. df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수를 출력하라.
`lst = [1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]`

```python
lst = [1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]

df[df.new_price.isin(lst)]
```

`df[df.new_price in lst]`으로 하면 시리즈에서 찾는 값이 너무 많다고 찾아내질 못하므로 `in` 연산자가 아니라 `isin` 연산자를 사용해야 함.

```python
len(df[df.new_price.isin(lst)])
```

데이터프레임의 갯수를 구할 때에는 `len`을 사용해주자.

## 03_Grouping

### 44. 데이터를 로드하고 상위 5개 컬럼을 출력하라

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df
```

### 45. 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
pd.DataFrame(df['host_name'].value_counts()).sort_index().head(5)
```

### 46. 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라.
### 빈도수 컬럼은 counts로 명명하라.

```python
df_cnt = pd.DataFrame(df['host_name'].value_counts()).sort_values('host_name', ascending = False)
df_cnt.rename(columns = {'':'host_name', 'host_name':"counts"}, inplace = True)

df_cnt
```

### 47. neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라

```python
pd.DataFrame(df.groupby(['neighbourhood_group', 'neighbourhood']).size())
```

**groupby 조건이 2개 이상일 경우 [] (리스트)를 써서 작업하자.**

### 48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
pd.DataFrame(df.groupby(['neighbourhood_group', 'neighbourhood']).size().groupby('neighbourhood_group').max())
```

다중 조건으로 그룹핑된 그룹에서 바로 max를 사용하면 한 칼럼을 대상으로 해야 한다며 적용이 안 됨.          
**따라서 다중 조건으로 그룹핑한 시리즈나 데이터프레임에서 다시 최댓값을 구하려는 칼럼을 그룹핑 한 다음 최댓값을 취해야 함.**

### 49. neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라

```python
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean','var','min','max'])
```

집계함수 여러 번에 사용할 때는 `데이터프레임.agg(['min','max'...etc])`, 이 방식을 사용해서 보여주면 됨.

### 50. neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라

```python
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean','var','min','max'])
```
