## 비행탑승 경험 만족도 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

### Question 92

Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
df92 = df[(df['Arrival Delay in Minutes'].isnull() == True)]
df92 = df92[['satisfaction', 'Class']].groupby(['Class', 'satisfaction'], as_index=False).size().pivot(index='Class',columns='satisfaction')
df92[(df92['size']['neutral or dissatisfied'] < df92['size']['satisfied'])]
```

## 의료 비용 예측 데이터

### Question 94

흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
df_smoker = df[(df.smoker == 'yes')]
df_nsmoker = df[(df.smoker == 'no')]

df_smoker[df_smoker.charges >= df_smoker.charges.quantile(0.9)].charges.mean() - df_nsmoker[df_nsmoker.charges >= df_nsmoker.charges.quantile(0.9)].charges.mean()
```

### Question 95

## 대학원 입학 가능성 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv")
df.head()
```

### Question 96

Serial No. 컬럼을 제외하고 ‘Chance of Admit’을 종속변수, 나머지 변수를 독립변수라 할때, 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을 출력하라 (시드값에 따라 순서는 달라질수 있음)

ml 파트부터 다시 풀어보기

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
df = df.drop('Serial No.', axis = 1)

y = df['Chance of Admit']
x = df[[ 'GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research']]

ml = RandomForestRegressor()
ml.fit(x,y)

result=pd.DataFrame({'importance':ml.feature_importances_},x.columns).sort_values('importance',ascending=False)
```

```python
result
```

## 레드 와인 퀄리티 예측 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/train.csv")
df.head()
```

### Question 97

quality 값이 3인 그룹과 8인 데이터그룹의 각 컬럼별 독립변수의 표준편차 값의 차이를 구할때 그값이 가장 큰 컬럼명을 구하여라

```python
answer = (df.loc[df.quality ==8].std() -df.loc[df.quality ==3].std()).sort_values().index[-1]
print(answer)
```

## 약물 분류 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv")
df.head()
```

### Question 98

남성들의 연령대별 (10살씩 구분 0-9세,10-19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

```python
def age_cg(x):
  return (x // 10) * 10

df['Age2'] = df.Age.apply(age_cg)
df2 = df[(df.Sex == 'M')]
df2[['Age2', 'Na_to_K']].groupby('Age2').mean()
```
