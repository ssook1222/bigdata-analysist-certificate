## 패키지 다 때려박기

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_test.csv')
```

#### 데이터셋 분리하기

```python
x_train_tr, x_train_ts, y_train_tr, y_train_ts = train_test_split(x_train, y_train, test_size = 0.33, random_state = 10)
```

### 데이터 전처리하기

```python
display(x_train_tr)
display(y_train_tr)
display(x_train_ts)
display(y_train_ts)
```

```python
x_train_tr[x_train_tr.isna() == True].count()
```

```python
x_train_tr.isna().sum()
```

결측치가 있는 열 제거

```python
x_train_tr.columns
```

```python
x_train_tr = x_train_tr[['ID', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
       'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']]

x_train_ts = x_train_ts[['ID', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
       'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']]

x_test = x_test[['ID', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
       'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']]
```

#### 라벨 인코딩

```python
x_train_tr.info()
```

```python
le = LabelEncoder()

x_train_tr['Gender'] = le.fit_transform(x_train_tr['Gender'])
x_train_tr['Customer Type'] = le.fit_transform(x_train_tr['Customer Type'])
x_train_tr['Type of Travel'] = le.fit_transform(x_train_tr['Type of Travel'])
x_train_tr['Class'] = le.fit_transform(x_train_tr['Class'])

x_train_ts['Gender'] = le.fit_transform(x_train_ts['Gender'])
x_train_ts['Customer Type'] = le.fit_transform(x_train_ts['Customer Type'])
x_train_ts['Type of Travel'] = le.fit_transform(x_train_ts['Type of Travel'])
x_train_ts['Class'] = le.fit_transform(x_train_ts['Class'])

x_test['Gender'] = le.fit_transform(x_test['Gender'])
x_test['Customer Type'] = le.fit_transform(x_test['Customer Type'])
x_test['Type of Travel'] = le.fit_transform(x_test['Type of Travel'])
x_test['Class'] = le.fit_transform(x_test['Class'])
```

```python
y_train_tr['satisfaction'] = le.fit_transform(y_train_tr['satisfaction'])
y_train_ts['satisfaction'] = le.fit_transform(y_train_ts['satisfaction'])
y_test['satisfaction'] = le.fit_transform(y_test['satisfaction'])
```

```python
display(x_train_tr)
display(x_train_ts)
```

차이가 크게 나는 행이 있음. 정규화 진행 필요

#### Min-Max 정규화 진행

```python
MinMax = MinMaxScaler()
```

```python
x_train_tr_data = MinMax.fit_transform(x_train_tr)
x_train_ts_data = MinMax.fit_transform(x_train_ts)
x_test_data = MinMax.fit_transform(x_test)

x_train_tr = pd.DataFrame(columns = x_train_tr.columns, data = x_train_tr_data)
x_train_ts = pd.DataFrame(columns = x_train_ts.columns, data = x_train_ts_data)
x_test = pd.DataFrame(columns = x_test.columns, data = x_test_data)
```

### 데이터 모델링 진행

```python
x_train_tr.info()
```

```python
rfc = RandomForestClassifier()
```

```python
x_train_tr = x_train_tr[:1000]
y_train_tr = y_train_tr[:1000]
```

```python
x_test = x_test[:1000]
```

```python
model = rfc.fit(x_train_tr, y_train_tr)
```

```python
predict1 = model.predict(x_train_ts)
predict2 = model.predict(x_test)
```

```python
y_test = y_test[:1000]
predict1 = predict1[:1000]
```

```python
y_test
```

```python
print(accuracy_score(predict1[:,1], y_test['satisfaction']))
```

### 제출하기

```python
predict2
```

```python
def satisfact(x):
  if x == 0:
    return 'neutral or dissatisfied'
  else:
    return 'satisfied'
```

```python
submit = pd.DataFrame({'ID':y_test['ID'], 'satisfaction':predict2[:,1]})
```

```python
submit['satisfaction'] = submit.satisfaction.apply(lambda x:satisfact(x))
```

```python
submit.to_csv('result.csv')
```
