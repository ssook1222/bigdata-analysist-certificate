## 당뇨여부판단 데이터

### 패키지 다 때려박기

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 다 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/y_test.csv')
```

### 데이터 전처리하기

```python
display(x_train)
display(y_train)
display(x_test)
display(y_test)
```

#### 데이터 결측치 확인

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

결측치 없음. 전부 다 수치형인지 확인.

```python
print(x_train.info())
print(x_test.info())
```

전부 수치형. 열 간의 규모 차이 확인.

```python
display(x_train)
display(x_test)
```

차이가 꽤 큼.      
최대-최소 정규화 진행 및 각 데이터 프레임 ID 제거

```python
print(x_train.columns)
```

```python
x_train = x_train[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
       'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
x_test = x_test[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
       'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
min_max_scaler = MinMaxScaler()
```

```python
x_train_column = x_train.columns

x_train_data = min_max_scaler.fit_transform(x_train)
x_test_data = min_max_scaler.fit_transform(x_test)

x_train = pd.DataFrame(data = x_train_data, columns = x_train_column)
x_test = pd.DataFrame(data = x_test_data, columns = x_train_column)
```

```python
display(x_train)
display(x_test)
```

### 데이터 모델링 진행

```python
dct = DecisionTreeClassifier()
```

```python
model = dct.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
display(y_test)
```

```python
print(accuracy_score(predict[:,1], y_test['Outcome']))
```

### 제출하기

```python
submit = pd.DataFrame(data = predict, columns = y_test.columns)
```
