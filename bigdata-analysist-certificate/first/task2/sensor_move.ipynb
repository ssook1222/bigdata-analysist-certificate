### 데이터 다 가져오기

### 모델 생성

```python
dtc = DecisionTreeClassifier(max_depth=3, max_leaf_nodes = 2, max_features= 3)
```

```python
model = dtc.fit(x_train_real, y_train)
```

```python
predict = model.predict(x_test_real)
```

```python
print(accuracy_score(predict[:,1], y_test['pose']))
```

### 제출하기
