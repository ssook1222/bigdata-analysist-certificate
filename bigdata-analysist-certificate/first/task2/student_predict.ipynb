## 학생성적 예측 데이터

### 패키지 다 때려박기

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.tree import DecisionTreeRegressor
```

#### 회귀 모델은 아래와 같은 걸로 평가 지표 사용

```python
from sklearn.metrics import mean_absolute_error
```

### 데이터 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_test.csv')
```

### 데이터 전처리

#### 데이터 프레임 확인

```python
display(x_train)
display(x_test)
```

#### 결측치 확인

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

둘 다 없음. ML을 위한 인코딩 진행

```python
print(x_train.info())
print(x_test.info())
```

#### 라벨 인코딩 진행

```python
le = LabelEncoder()
```

```python
columns = x_train.select_dtypes(exclude = [np.int64, np.float64]).columns
```

```python
for column in columns:
  x_train[column] = le.fit_transform(x_train[column])
  x_test[column] = le.fit_transform(x_test[column])
```

#### 필요없는 칼럼 제거

```python
x_train = x_train[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus',
       'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime',
       'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities',
       'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime',
       'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2']]

x_test = x_test[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus',
       'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime',
       'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities',
       'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime',
       'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2']]
```

### 모델링 진행

```python
dcr = DecisionTreeRegressor()
```

```python
model = dcr.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
print(mean_absolute_error(predict[:,1], y_test['G3']))
```
