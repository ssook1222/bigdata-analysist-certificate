## 수질 음용성 여부 데이터

### 패키지 다 가져오기

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 불러오기

```python
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_train.csv")
x_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_test.csv")
y_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_test.csv")
```

### 데이터 전처리

```python
display(x_train)
display(y_train)
display(x_test)
display(y_test)
```

#### 결측치 제거
결측치가 있으므로 중간값을 사용하여 결측치를 대체함.

```python
imputer_mid = SimpleImputer(strategy='median')

x_train_columns = x_train.columns
x_test_columns = x_test.columns

imputer_mid.fit(x_train)
x_train_data = imputer_mid.transform(x_train)

imputer_mid.fit(x_test)
x_test_data = imputer_mid.transform(x_test)

x_train = pd.DataFrame(columns = x_train_columns, data=x_train_data)
x_test = pd.DataFrame(columns = x_test_columns, data=x_test_data)
```

#### 데이터 타입 확인 후 인코딩 진행

```python
x_train.info()
x_test.info()
```

할 필요가 없음.

#### 정규화 진행

특정한 열의 데이터 크기 차이가 꽤 나므로 이 부분 정규화 따로 진행.

```python
min_max = MinMaxScaler()
```

```python
x_train_data = min_max.fit_transform(x_train)
x_train = pd.DataFrame(columns = x_train_columns, data = x_train_data)
display(x_train)

x_test_data = min_max.fit_transform(x_test)
x_test = pd.DataFrame(columns = x_test_columns, data = x_test_data)
display(x_test)
```

### 데이터 모델링 진행

```python
dtc = DecisionTreeClassifier()
```

```python
model = dtc.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
print(accuracy_score(predict[:,1], y_test['Potability']))
```

### 제출하기

```python
submit = pd.DataFrame(columns = y_test.columns, data = predict)
```
