### Question 22

quantity컬럼 값이 3인 데이터를 추출하여 index를 1부터 정렬하고 첫 5행을 출력하라

```python
df22 = df[(df.quantity == 3)].reset_index()
df22 = df22[['quantity', 'item_name','choice_description', 'item_price']]
df22.index += 1
df22.head(5)
```

`df22.index += 1`                
→ 이러한 방식으로 index 직접 접근 가능

### Question 24

원본데이터에서 item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

```python
def chg(x):
  x_raw = list(x)[1:]
  return float(''.join(x_raw))

df24 = df
df24['new_price'] = df24.item_price.apply(chg)
```

```python
df24
```

### Question 25

new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

```python
df25 = df24[(df24.new_price <=5 )].count().values[0]
print(df25)
```
