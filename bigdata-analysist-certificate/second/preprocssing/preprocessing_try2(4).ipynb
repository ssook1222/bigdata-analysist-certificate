```python
import pandas as pd
```

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)
```

```python
def chg(x):
  x_raw = list(x)[1:]
  return float(''.join(x_raw))

df2 = df
df2['new_price'] = df2.item_price.apply(chg)
```

### Question 31

df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라

```python
df31 = df.sort_values('new_price', ascending = False).reset_index(drop = True)
display(df31)
```

### Question 32

```python
display(df2[(df2.item_name == 'Steak Salad') | (df2.item_name == 'Bowl')])
```

### Question 33

df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
df33 =df2[(df2.item_name == 'Steak Salad') | (df2.item_name == 'Bowl')]
display(df33.drop_duplicates('item_name', keep='first'))
```

### Question 40

df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력하라

```python
df.choice_description
```

```python
cnt = 0

def check(x):
  global cnt

  if 'Vegetables' not in x:
    cnt += 1
```

```python
df.choice_description.apply(check)
print(cnt)
```
