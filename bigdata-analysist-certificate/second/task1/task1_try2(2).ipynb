### Question 14

각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자 수(subcnt) 의 증가량을 구하여라

```python
channel[(channel.ct >= pd.to_datetime('2021-10-03 03:00:00')) & (channel.ct <= pd.to_datetime('2021-11-01 15:00:00'))]
```

```python
first = channel[(channel.ct >= pd.to_datetime('2021-10-03 03:00:00'))].drop_duplicates('channelname', keep = 'first')['subcnt'].values
second = channel[(channel.ct >= pd.to_datetime('2021-11-01 15:00:00'))].drop_duplicates('channelname', keep = 'last')['subcnt'].values
inc = []
for i in range(len(first)):
  inc.append(second[i] - first[i])
ans = pd.DataFrame({'channelname':channel[(channel.ct >= pd.to_datetime('2021-10-03 03:00:00'))].drop_duplicates('channelname', keep = 'first')['channelname'].tolist(), 'inc': inc})
display(ans)
```

### Question 16

각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라

```python
df16 = video.sort_values('ct', ascending = True).drop_duplicates('videoname', keep = 'first')

def change_dt(x):
  raw = str(x).split()
  return raw[0]

df16['start_date'] = df16.ct.apply(change_dt)
df16 = df16[['start_date','videoname']]
display(df16)
```

### Question 17

“공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.

### Question 18

video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

### Question 19

2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
first = video[((video.ct >= pd.to_datetime('2021-11-01 00:00:00')) & (video.ct <= pd.to_datetime('2021-11-01 15:00:00')))].sort_values('ct').drop_duplicates('videoname', keep = 'first')
second = video[((video.ct >= pd.to_datetime('2021-11-01 00:00:00')) & (video.ct <= pd.to_datetime('2021-11-01 15:00:00')))].sort_values('ct', ascending = False).drop_duplicates('videoname', keep = 'first')
```

```python
first = first[['videoname', 'viewcnt']].sort_values('videoname').reset_index(drop = True)
second = second[['videoname', 'viewcnt']].sort_values('videoname').reset_index(drop = True)
```

```python
video_name = first['videoname'].to_list()
data = list((second.viewcnt - first.viewcnt).values)
```

```python
ans = pd.DataFrame({'videoname':video_name, 'viewcnt':data})
display(ans)
```
