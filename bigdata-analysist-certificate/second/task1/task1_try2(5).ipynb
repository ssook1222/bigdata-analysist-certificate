## 전세계 행복도 지표 데이터

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/happy2/happiness.csv',encoding='utf-8')
df.head()
```

### Question 41

데이터는 2018년도와 2019년도의 전세계 행복 지수를 표현한다. 각년도의 행복랭킹 10위를 차지한 나라의 행복점수의 평균을 구하여라

```python
df2018 = df[df.년도 == 2018].sort_values('행복랭킹', ascending = True).점수.values[9]
df2019 = df[df.년도 == 2019].sort_values('행복랭킹', ascending = True).점수.values[9]

print((df2018 + df2019)/2)
```

### Question 42

데이터는 2018년도와 2019년도의 전세계 행복 지수를 표현한다. 각년도의 행복랭킹 50위이내의 나라들의 각각의 행복점수 평균을 데이터프레임으로 표시하라

```python
mean2018 = df[df.년도 == 2018].점수.head(50).mean()
mean2019 = df[df.년도 == 2019].점수.head(50).mean()

means = [[2018, mean2018],[2019, mean2019]]
column = ['년도', '점수']
```

```python
pd.DataFrame(data = means, columns = column)
```

### Question 45
2019년도 데이터들만 추출하여 각변수간 상관계수를 구하고 내림차순으로 정렬한 후 상위 5개를 데이터 프레임으로 출력하라. 컬럼명은 v1,v2,corr으로 표시하라

### Question 47

2019년 데이터를 추출하고 해당데이터의 상대 GDP 평균 이상의 나라들과 평균 이하의 나라들의 행복점수 평균을 각각 구하고 그 차이값을 출력하라

### Question 48

각년도의 부패에 대한인식을 내림차순 정렬했을때 상위 20개 국가의 부패에 대한인식의 평균을 구하여라

### Question 49

2018년도 행복랭킹 50위 이내에 포함됐다가 2019년 50위 밖으로 밀려난 국가의 숫자를 구하여라

! 다시 한 번 풀어보기

`set` 자료구조를 사용해서 아래와 같이 조건에 맞는 걸 차집합으로 해 갖구 풀 수 있음.

```python
result = set(df[(df.년도==2018) & (df.행복랭킹 <=50)].나라명)  -set(df[(df.년도==2019) & (df.행복랭킹 <=50)].나라명)
answer = len(result)
print(answer)
```

### Question 50

2018년, 2019년 모두 기록이 있는 나라들 중 년도별 행복점수가 가장 증가한 나라와 그 증가 수치는?

```python
df.나라명.value_counts() # 2018년, 2019년 기록 확인
```

```python
count = df.나라명.value_counts()
```

```python
lst = count[count >= 2].index
```

```python
df50 = df[(df.나라명.isin(lst))]
```

```python
df50_2018 = df50[['나라명', '년도', '점수']]
df50_2018 = df50_2018[(df50_2018.년도==2018)]
```

```python
df50_2019 = df50[['나라명', '년도', '점수']]
df50_2019 = df50_2019[(df50_2019.년도==2019)]
```

```python
df50_raw = pd.merge(df50_2018, df50_2019, how='inner', on='나라명')
```

```python
df50_raw['diff'] = df50_raw['점수_y']-df50_raw['점수_x']
```

```python
print(df50_raw.sort_values('diff', ascending = False).head(1).나라명.values[0])
```
