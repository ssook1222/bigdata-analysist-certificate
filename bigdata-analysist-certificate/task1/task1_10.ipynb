## 센서데이터 동작유형 분류 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv")
df.head()
```

### Question 100

pose값에 따른 각 motion컬럼의 중간값의 가장 큰 차이를 보이는 motion컬럼은 어디이며 그값은?

```python
t = df.groupby('pose').median().T

abs(t[0] - t[1]).sort_values(ascending = False).head(2).index
```

T를 붙이면 행렬 전치 가능

### Question 101

## 넷플릭스 주식 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")
df.head()
```

### Question 103

매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라

```python
import datetime as dt
```

```python
df['Date'] = pd.to_datetime(df.Date)
df103 = df[(df['Date'].dt.month == 5)]
df103['Date'] = pd.to_datetime(df103.Date)
df103['Date'] = df103.Date.dt.strftime('%Y-%m')
df103[['Date', 'Open']].groupby('Date').mean()
```

`strftime('%Y-%m'))` : 년도 - 월로 이렇게 DateTime 원하는 대로 양식 세팅 가능

## nba 선수 능력치 데이터

```python
pd.set_option('display.max_columns',50)
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()
```

### Question 104

Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

```python
round(df[(df.Tm == 'TOR')].mean().Age, 4)
```

### Question 105

```python
df[(df.Age!=df.Age.min)].Pos.value_counts().head(1).index
```

### Question 106

선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라

### Question 107

PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

```python
df[['PTS', 'Pos']].groupby('Pos', as_index = False).mean()
```

### Question 109

G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라

```python
df[['G', 'Tm']].groupby('Tm', as_index = False).max().mean().G
```

### Question 110

Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

```python
df[(df.Tm == 'MIA') & ((df.Pos == 'C') | (df.Pos == 'PF'))].MP.mean()
```

### Question 111

전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

```python
df[(df.G >= df.G.mean() * 1.5)]['3P'].mean()
```

### Question 112

Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

```python
df112_up = df[(df.Age >= df.Age.mean())]
df112_down = df[(df.Age < df.Age.mean())]

df112_up.G.mean() - df112_down.G.mean()
```

### Question 113
