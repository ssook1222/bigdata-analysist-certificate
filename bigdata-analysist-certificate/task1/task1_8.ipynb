## 기온 강수량 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/weather/weather2.csv")
df.head()
```

```python
import datetime as dt

df.time = pd.to_datetime(df.time)
```

### Question 81

여름철(6월,7월,8월) 이화동이 수영동보다 높은 기온을 가진 시간대는 몇개인가?

```python
df_81 = df[df['time'].dt.month.isin([6,7,8])]
df_81['time'] = pd.to_datetime(df_81.time)
```

```python
df_81[(df_81.이화동기온 > df_81.수영동기온)].time.dt.hour.size
```

### Question 82

이화동과 수영동의 최대강수량의 시간대를 각각 구하여라

## 서비스 이탈예측 데이터

```python
import pandas as pd
#데이터 로드
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/train.csv")
df.head()
```

### Question 83

남성 이탈(Exited)이 가장 많은 국가(Geography)는 어디이고 이탈 인원은 몇명인가?

```python
df[(df.Gender=='Male')].groupby('Geography').sum().Exited.sort_values(ascending=False).head(1)
```

### Question 84

카드를 소유(HasCrCard ==1)하고 있으면서 활성멤버(IsActiveMember ==1) 인 고객들의 평균 나이를 소숫점이하 4자리까지 구하여라?

```python
round(df[(df.HasCrCard == 1) & (df.IsActiveMember == 1)].Age.mean() , 4)
```

### Question 85

Balance 값이 중간값 이상을 가지는 고객들의 CreditScore의 표준편차를 소숫점이하 3자리까지 구하여라

```python
round(df[(df.Balance >= df.Balance.median())].CreditScore.std(), 3)
```

## 성인 건강검진 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/train.csv")
df.head()
```

### Question 86

수축기혈압과 이완기 혈압기 수치의 차이를 새로운 컬럼(‘혈압차’) 으로 생성하고, 연령대 코드별 각 그룹 중 ‘혈압차’ 의 분산이 5번째로 큰 연령대 코드를 구하여라

```python
df['혈압차'] = df['수축기혈압'] - df['이완기혈압']
df[['혈압차', '연령대코드(5세단위)']].groupby('연령대코드(5세단위)').var().혈압차.sort_values(ascending = False).head(5).index[4]
```

### Question 87

비만도를 나타내는 지표인 WHtR는 허리둘레 / 키로 표현한다. 일반적으로 0.58이상이면 비만으로 분류한다. 데이터중 WHtR 지표상 비만인 인원의 남/여 비율을 구하여라

```python
df['WHtR'] = df.허리둘레/ df['신장(5Cm단위)']
df[(df.WHtR >= 0.58) & (df.성별코드 == 'M')].index.size / df[(df.WHtR >= 0.58) & (df.성별코드 == 'F')].index.size
```

## 자동차 보험가입 예측데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/train.csv")
df.head()
```

### Question 88

Vehicle_Age 값이 2년 이상인 사람들만 필터링 하고 그중에서
Annual_Premium 값이 전체 데이터의 중간값 이상인 사람들을 찾고, 그들의 Vintage값의 평균을 구하여라

```python
df_88 = df[(df.Vehicle_Age == '> 2 Years')]
df_88[(df_88.Annual_Premium.median() <= df_88.Annual_Premium)].Vintage.mean()
```

### Question 89

vehicle_age에 따른 각 성별(gender)그룹의 Annual_Premium값의 평균을 구하여 아래 테이블과 동일하게 구현하라

나중에 다시 한 번 풀어보기!

```python
df_89 = df.groupby(['Gender','Vehicle_Age'],as_index = False)['Annual_Premium'].mean()
df_89.pivot(index='Vehicle_Age',columns='Gender',values='Annual_Premium')
```
