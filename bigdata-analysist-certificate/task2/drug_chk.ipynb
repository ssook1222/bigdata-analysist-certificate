## 약물 분류 데이터

### 패키지 다 때려박기

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_test.csv')
```

### 데이터 전처리

```python
display(x_train)
display(x_test)
```

#### 결측치 확인

```python
x_train.isna().sum()
```

없으므로 그대로 진행

#### 라벨 인코딩 진행

```python
x_train.info()
```

```python
le = LabelEncoder()
```

```python
x_train['Sex'] = le.fit_transform(x_train['Sex'])
x_train['BP'] = le.fit_transform(x_train['BP'])
x_train['Cholesterol'] = le.fit_transform(x_train['Cholesterol'])
```

```python
x_test['Sex'] = le.fit_transform(x_test['Sex'])
x_test['BP'] = le.fit_transform(x_test['BP'])
x_test['Cholesterol'] = le.fit_transform(x_test['Cholesterol'])
```

칼럼별 크기 규모 확인

```python
display(x_train)
display(x_test)
display(y_train)
```

많이 차이 안 나므로 그대로 진행

### 데이터 모델링

```python
dt = DecisionTreeClassifier()
```

```python
model = dt.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
y_test
```

```python
print(accuracy_score(predict[:,1], y_test['Drug']))
```

```python
y_test['Drug']
```

```python
predict
```

### 제출하기

```python
submit = pd.DataFrame(data = predict, columns = y_test.columns)
```
