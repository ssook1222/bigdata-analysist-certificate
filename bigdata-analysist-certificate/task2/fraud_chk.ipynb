## 사기회사 분류 데이터

### 패키지 다 때려박기

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 다 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/y_test.csv')
```

### 데이터 전처리하기

```python
display(x_train)
display(y_train)
display(x_test)
display(y_test)
```

#### 결측치 확인

```python
x_train.isna().sum()
```

```python
x_test.isna().sum()
```

애매하게 1개씩 있음.    
imputer로 결측치 제거

```python
si = SimpleImputer(strategy='most_frequent')
```

```python
x_train_columns = x_train.columns
x_test_columns = x_test.columns

x_train_data = si.fit_transform(x_train)
x_test_data = si.fit_transform(x_test)
```

```python
x_train = pd.DataFrame(data = x_train_data, columns = x_train_columns)
x_test = pd.DataFrame(data = x_test_data, columns = x_test_columns)
```

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

#### 라벨 인코딩 진행

```python
x_train.info()
```

```python
le = LabelEncoder()
```

```python
# 데이터 프레임 전체에 라벨 인코딩 진행

for column in x_train.columns:
    x_train[column] = le.fit_transform(x_train[column])

for column in x_test.columns:
    x_test[column] = le.fit_transform(x_test[column])
```

라벨링 이후 값의 규모가 비슷해짐.

#### 데이터 모델링 진행

```python
dcf = DecisionTreeClassifier()
```

```python
model = dcf.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
print(accuracy_score(predict[:, 1], y_test['Risk']))
```

### 제출하기

```python
submit = pd.DataFrame(columns = y_test.columns, data = predict)
```
