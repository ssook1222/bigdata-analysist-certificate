## 단일 표본 t 검정

### Question 1.

100명의 키 정보가 들어 있는 데이터가 있다.데이터가 정규성을 만족하는지 확인하라.     
그리고 평균키는 165라 판단할수 있는지 귀무가설과 대립가설을 설정한 후 유의수준 5%로 검정하라.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/height2.csv')
```

```python
from scipy.stats import ttest_1samp # 단일표본 t검정에 사용
from scipy.stats import shapiro # 정규분포를 가장 엄격하게 검정하며 데이터수가 상대적으로 적을 때 사용
from scipy.stats import wilcoxon # 윌콕슨 부호 순위합 검정
```

```python
# 샤피로를 통한 정규성 검정 진행
static, pvalue = shapiro(df)
pvalue
```

p-value가 0.013552471995353699 이므로 귀무가설을 기각(= 정규성을 만족하지 않음).      
따라서 비모수 검정(= 윌콕슨 순위 부호 검정)을 진행.



```python
#윌콕슨 부호순위 검정
static, pvalue = wilcoxon(df['height']-165)
```

pvalue는 0.04880534424649582이므로, 귀무가설을 기각하고 대립가설을 채택한다.      
이때 귀무가설은 `100명 키의 평균은 165이다.` 인데, 위의 결과처럼             
해당 가설을 기각하고 대립가설을 채택하게 되었으므로 `100명 키의 평균은 165가 아니다.`

## 등분산 검정

### Question 2.

두개 학급의 시험성적에 대한 데이터이다. 그룹간 등분산 검정을 시행하라

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy2.csv')
df.head()
```

```python
# 등분산 검정에서 사용하는 라이브러리
from scipy.stats import bartlett
from scipy.stats import fligner
from scipy.stats import levene
```

```python
a = df[df['class'] =='A'].score
b = df[df['class'] =='B'].score

print(bartlett(a,b))

print(fligner(a,b,center='median')) #default
print(fligner(a,b,center='mean'))

print(levene(a,b, center='median')) #default
print(levene(a,b,center='mean'))
```

등분산검정의 방법은 3가지가 있다.       
p value값은 5% 유의수준일 때, 0.05보다 작은 경우 `"각 그룹은 등분산이다"`라는 귀무가설을 기각하면 된다.
위의 결과를 보면 모두 0.05보다 크므로 귀무가설을 기각할 수 없음을 알 수 있다.    
즉, `두 개 학급의 시험 성적은 등분산이 아니`라는 것을 의미한다.

### Question3

두 개 학급의 시험성적에 대한 데이터이다. 그룹 간 등분산 검정을 시행하라.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy3.csv')
df.head()
```

```python
from scipy.stats import bartlett
from scipy.stats import fligner
from scipy.stats import levene
```

```python
a = df[df['class'] =='A'].score
b = df[df['class'] =='B'].score

print(bartlett(a,b))

print(fligner(a,b,center='median')) #default
print(fligner(a,b,center='mean'))

print(levene(a,b, center='median')) #default
print(levene(a,b,center='mean'))
```

fligner, levene 검정 결과 pvalue는 0.05보다 작은데, Bartlett 검정의 pvalue는 0.05보다 크다.         
이 때 Bartlett 검정은 이상치에 영향을 받기 쉬우므로   
fligner, levene 검정의 pvalue가 0.05보다 작게 나왔으므로 해당 결과를 채택하면,  `"각 그룹은 등분산이다"`라는 귀무가설을 채택할 수 있다.   
즉, 두 학급의 시험 성적 데이터는 등분산이다.
